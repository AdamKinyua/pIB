```python
import numpy as np
from numpy import sin, pi
import matplotlib.pyplot as plt 
```

```python
def coeff_n(n,a):
    #first sine variables
    x1 = (n/a-1)*pi
    x2 = 1/(n/a-1)
    #second sine variables 
    y1 = (n/a+1)*pi
    y2 = 1/(n/a+1)
    Cn = 1/(np.sqrt(a)*np.pi)*(sin(x1*x2)-sin(y1*y2)) #FOR THIS SOLUTION, N CANNOT BE EQUAL TO 1
    return Cn
coeff_n(n=2, a = 1)
```

```python
#Definitions
# a                   - 
# L                   - 
# x                   - 
# dx                  - 
# Ψn = sin(n*pi*x/aL) -
# Ψm = sin(pi*x/L)    -

def integral_psi_m_psi_n_dx(n,x,dx,L,a):
    psi_list = []
    x_list = []
    while x<=a: 
        psi = 2/(L*np.sqrt(a))*sin(n*pi*x/(a*L))*sin(pi*x/L)*dx
        x += dx*a
        psi_list.append(psi)
        x_list.append(x)
    plt.plot(x_list,psi_list)
    plt.xlabel("box length L")
    plt.ylabel("psi")
 

integral_psi_m_psi_n_dx(n=1.5, x=0, dx=0.001,L=1.5,a=1)
```
