```python
import matplotlib.pyplot as plt
import numpy as np
'''
First Section: [3] - [9]
Hardconding by guessing our eigen values E

Second Section [10]
Accurately finding our first eigen value

Third Section [11]
Finding Subsequent Eigen Values 
'''
```

```python
# defining h-bar and  the mass of the particle 
h=1
m=1
```

```python
#defining the length of the box and the potential inside the box
a=1
V=0
```

```python
# Defining our wavefunction and the first derivative; these two values have to be different
psi=0
dpsi=1
```

```python
#Remember x starts from 0 to a and we can increase it gradually by dx
x=0
dx=0.001*a
```

```python
#Storing the xvalues and the vales of our wavefunction as x changes 
xlist=[]
psilist=[]
```

```python
#Defining our eigenvalue; here, since we are using the shooting method, we will adjust the  values of E until we converge
E=4.9
```

```python
#Solving our wavefunction for all the  values of values of x
while x<=a:
    ddpsi=2*m/h**2*(V-E)*psi
    dpsi=dpsi+ddpsi*dx
    psi=psi+dpsi*dx
    x=x+dx
    xlist.append(x)
    psilist.append(psi)
    plt.plot(xlist,psilist)
    plt.xlabel("box length L")
    plt.ylabel("psi")
    plt.grid(visible=None, which='major', axis='both', color='r', linestyle='-', linewidth=1) 
    #which: major, minor, both
```

```python
#Accurately evaluating our first eigen value
E=0
dE=0.01
dx=0.001*a
psi=10 # redefining psi with random value

while abs(psi)>0.001: 
    psi=0
    dpsi=1
    E=E+dE
    x=0
    xlist=[]
    psilist=[]

    while x<=a:
        ddpsi=2*m/h**2*(V-E)*psi
        dpsi=dpsi+ddpsi*dx
        psi=psi+dpsi*dx
        x=x+dx
        xlist.append(x)
        psilist.append(psi)
plt.plot(xlist,psilist)
plt.xlabel("box length L")
plt.ylabel("psi")
plt.grid(visible=None, which='major', axis='both', color='r', linestyle='-', linewidth=1) 

print(f'First Eigen Value: {E}')
```

```python
# Finding subsequent Eigen  Values 
E=0
dE=0.01
dx=0.001*a
counter = 1
nmax=5
Eigen_Values=[]
EigenFunctions = []
EigenFunctionsXlist=[]
while counter<=nmax:
    psi=10 # ANy random number can do

    while abs(psi)>0.001: 
        psi=0
        dpsi=1
        E=E+dE
        x=0
        xlist=[]
        psilist=[]

        while x<=a:
            ddpsi=2*m/h**2*(V-E)*psi
            dpsi=dpsi+ddpsi*dx
            psi=psi+dpsi*dx
            x=x+dx
            xlist.append(x)
            psilist.append(psi)
            
    Eigen_Values.append(E)
    EigenFunctions.append([psilist]) 
    EigenFunctionsXlist.append(xlist)
    counter+=1
    E=E*1.1 #Avoids getting the same  eigen value twice
    
plt.plot(xlist,psilist)
plt.xlabel("box length L")
plt.ylabel("psi")
#plt.grid(visible=None, which='major', axis='both', color='r', linestyle='-', linewidth=1) 
print(f'List of EigenValues: {Eigen_Values}')
```

```python
counter=1
while counter<=nmax:
    plt.plot(xlist,EigenFunctions[counter-1][0])
    plt.xlabel("box length L")
    plt.ylabel("psi")
    counter+=1
    #THe plot is wrong as the wavefunction have a the same slope at the beginning and probability density will vary, so we normalize in [15]
```

```python
#Normalization of wavefunction
#Calculating the sum over the absolute square of the wavefunction and we multiply it withour increment
#Its basically the integral of our wavefunction 
np.dot(EigenFunctions[2][0], EigenFunctions[2][0])*dx #Normalization integral
# THis gives us the norm for our third wavefunction as python counts from zero
#Now we need to divide wavefunctions by individual norm
#TO esnure every wavefunction has the same maximum; this is true; prefactor A SQRT(2/A)
```

```python
counter=1
while counter<=nmax:
    plt.plot(xlist,EigenFunctions[counter-1][0]/np.sqrt(np.dot(EigenFunctions[counter-1][0], EigenFunctions[counter-1][0])*dx))
    plt.xlabel("box length L")
    plt.ylabel("psi")
    counter+=1
```

```python
#ANother way of doing it 
counter = 1
while counter<=nmax:
    norm=np.dot(EigenFunctions[counter-1][0],EigenFunctions[counter-1][0])*dx
    EigenFunctions[counter-1][0][:] = [x/np.sqrt(norm) for x in EigenFunctions[counter-1][0]]
    counter+=1
```
